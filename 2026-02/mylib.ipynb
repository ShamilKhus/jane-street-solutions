From the original problem, I noticed that `sqrt(a + 2) / a` has to be an integer. That immediately leaves only two reasonable values for `a`: `a = 2` or `a = 1/4`.

After looking a little more carefully at the other expressions, it becomes clear that the workable choice is

- `a = 1/4`
- `c = 1/2`

After that, the only thing left is to test the integer options for `b` from `-3` to `3`, and only `b = - 3` works.


```python
import math
from fractions import Fraction
import matplotlib.pyplot as plt
```

```python
def as_frac(x) -> Fraction:
    if isinstance(x, Fraction):
        return x
    if isinstance(x, int):
        return Fraction(x, 1)
    if isinstance(x, float):
        return Fraction(x).limit_denominator()
    raise TypeError(f"Unsupported type for rational conversion: {type(x)}")


def as_int(x) -> int:
    if isinstance(x, int):
        return x
    if isinstance(x, Fraction) and x.denominator == 1:
        return int(x.numerator)
    if isinstance(x, float) and x.is_integer():
        return int(x)
    raise TypeError("b must be an integer.")


def _cuberoot_real(x: float) -> float:
    return math.copysign(abs(x) ** (1 / 3), x)


def pow_rational_safe(base, exp: int, max_abs_exp: int = 60):
    exp = as_int(exp)
    baseF = as_frac(base)

    if baseF == 0 and exp < 0:
        raise ZeroDivisionError("0 cannot be raised to a negative power")

    if abs(exp) <= max_abs_exp:
        if exp >= 0:
            return baseF ** exp
        return Fraction(1, 1) / (baseF ** (-exp))

    return float(baseF) ** exp


def _normalize(v):
    if isinstance(v, Fraction) and v.denominator == 1:
        return int(v.numerator)
    return v


def _val_to_str(v) -> str:
    if v is None:
        return "ERR"
    if isinstance(v, Fraction):
        if v.denominator == 1:
            return str(v.numerator)
        return f"{v.numerator}/{v.denominator}"
    if isinstance(v, float):
        if not math.isfinite(v):
            return str(v)
        return f"{v:.6g}"
    return str(v)
```

```python
def build_cells(a, b, c):
    aF = as_frac(a)
    cF = as_frac(c)
    bI = as_int(b)

    a_float = float(aF)
    c_float = float(cF)

    ab = pow_rational_safe(aF, bI)
    cb = pow_rational_safe(cF, bI)
    twob = pow_rational_safe(Fraction(2, 1), bI)

    return {
        (0, 4):  (r"$6c-4b$", lambda: 6*cF - 4*bI),
        (1, 7):  (r"$8-b$", lambda: 8 - bI),
        (2, 1):  (r"$\frac{a^b-4}{6c+1}$", lambda: (ab - 4) / (6*cF + 1)),
        (2, 3):  (r"$\frac{b+c}{c-1}$", lambda: (bI + cF) / (cF - 1)),
        (2, 5):  (r"$b^2-\frac{b}{c}$", lambda: Fraction(bI*bI, 1) - Fraction(bI, 1) / cF),
        (2, 7):  (r"$\frac{\sqrt{30+a}}{c}$", lambda: math.sqrt(30 + a_float) / c_float),
        (2,10):  (r"$\frac{a+b}{c-3a}$", lambda: (aF + bI) / (cF - 3*aF)),
        (3, 4):  (r"$\frac{b-3a}{a-c}$", lambda: (bI - 3*aF) / (aF - cF)),
        (3, 7):  (r"$8a-2b$", lambda: 8*aF - 2*bI),
        (3, 9):  (r"$\frac{b}{a-c}$", lambda: Fraction(bI, 1) / (aF - cF)),
        (3,11):  (r"$\frac{b+9}{\sqrt{c-a}}$", lambda: (bI + 9) / math.sqrt(float(cF - aF))),
        (4, 1):  (r"$\frac{18}{ac+1}$", lambda: Fraction(18, 1) / (aF*cF + 1)),
        (4, 5):  (r"$c^b$", lambda: cb),
        (4,10):  (r"$\frac{3+b^2}{\sqrt{3+2c}}$", lambda: (3 + bI*bI) / math.sqrt(3 + 2*c_float)),
        (5, 3):  (r"$\frac{b}{a^2-c^2}$", lambda: Fraction(bI, 1) / (aF*aF - cF*cF)),
        (5,12):  (r"$\frac{\sqrt{a+2}}{a}$", lambda: math.sqrt(float(aF + 2)) / float(aF)),
        (6, 2):  (r"$a^b-\frac{12}{a}$", lambda: (ab - Fraction(12, 1) / aF) if isinstance(ab, Fraction) else (ab - 12 / a_float)),
        (6, 4):  (r"$2c+\frac{c}{a}$", lambda: 2*cF + cF / aF),
        (6, 6):  (r"$4a-5b$", lambda: 4*aF - 5*bI),
        (6, 8):  (r"$c+2a$", lambda: cF + 2*aF),
        (6,10):  (r"$\frac{b}{9a-5c}$", lambda: Fraction(bI, 1) / (9*aF - 5*cF)),
        (7, 0):  (r"$\frac{b^3+2c}{b+2c}$", lambda: (bI**3 + 2*cF) / (bI + 2*cF)),
        (7, 9):  (r"$\frac{b}{a-1}$", lambda: Fraction(bI, 1) / (aF - 1)),
        (8, 2):  (r"$\frac{c-b}{2a}$", lambda: (cF - bI) / (2*aF)),
        (8, 7):  (r"$\frac{b}{a-c}$", lambda: Fraction(bI, 1) / (aF - cF)),
        (8,11):  (r"$\frac{b+c}{a-c}$", lambda: (bI + cF) / (aF - cF)),
        (9, 1):  (r"$\log_c(a)$", lambda: math.log(a_float) / math.log(c_float)),
        (9, 3):  (r"$\frac{c^2-b}{a}$", lambda: (cF*cF - bI) / aF),
        (9, 5):  (r"$(b-1)^2$", lambda: (bI - 1)**2),
        (9, 7):  (r"$\frac{\sqrt[3]{43-ac}}{a}$", lambda: _cuberoot_real(43 - float(aF*cF)) / a_float),
        (10,2):  (r"$\frac{b-a}{a-c}$", lambda: (bI - aF) / (aF - cF)),
        (10,4):  (r"$11-b$", lambda: 11 - bI),
        (10,6):  (r"$\frac{b-2a}{a-c}$", lambda: (bI - 2*aF) / (aF - cF)),
        (10,9):  (r"$\frac{c+3}{a}$", lambda: (cF + 3) / aF),
        (10,11): (r"$8c-\frac{b}{c}$", lambda: 8*cF - Fraction(bI, 1) / cF),
        (11,5):  (r"$b^2$", lambda: bI*bI),
        (12,7):  (r"$\frac{2^b+1}{ac}$", lambda: (twob + 1) / (aF*cF) if isinstance(twob, Fraction) else (twob + 1) / float(aF*cF)),
    }


def safe_eval(fn):
    try:
        return _normalize(fn())
    except Exception:
        return None


def cell_values(a, b, c):
    return {
        pos: safe_eval(fn)
        for pos, (_, fn) in build_cells(a, b, c).items()
    }
```

The choice that survives the rest of the structure is `a = 1/4`, and then `c = 1/2`.


```python
a = Fraction(1, 4)
c = Fraction(1, 2)


def all_integer_values(a, b, c):
    values = cell_values(a, b, c).values()
    return all(isinstance(v, int) for v in values)


[(b, all_integer_values(a, b, c)) for b in range(-3, 4)]
```

Only `b = -3` works, so the final parameters are

- `a = 1/4`
- `b = -3`
- `c = 1/2`


```python
def generate_solved_variables_13x13(a, b, c, outfile='solved_variables.png', show_expr=True, dpi=250):
    cells = build_cells(a, b, c)

    n = 13
    fig = plt.figure(figsize=(10, 10), dpi=dpi)
    ax = fig.add_subplot(111)
    ax.set_xlim(0, n)
    ax.set_ylim(0, n)
    ax.invert_yaxis()
    ax.set_aspect('equal')
    ax.axis('off')

    for i in range(n + 1):
        ax.plot([i, i], [0, n], linewidth=1, color='black')
        ax.plot([0, n], [i, i], linewidth=1, color='black')

    for (r, col), (tex_expr, fn) in cells.items():
        value = safe_eval(fn)
        value_str = _val_to_str(value)

        x = col + 0.5
        y = r + 0.5

        if show_expr:
            ax.text(x, y - 0.14, tex_expr, ha='center', va='center', fontsize=10)
            ax.text(x, y + 0.18, value_str, ha='center', va='center', fontsize=9)
        else:
            ax.text(x, y, value_str, ha='center', va='center', fontsize=10)

    plt.tight_layout(pad=0.2)
    plt.savefig(outfile, bbox_inches='tight')
    plt.close(fig)
    return outfile
```
